```python
# download packages from NLTK

import nltk
import string
import math
import operator
from nltk.corpus import brown, stopwords
from nltk.stem import LancasterStemmer
from nltk.probability import FreqDist
from collections import defaultdict, Counter


nltk.download('brown')
nltk.download('stopwords')
```

```python
#1. Load the Brown Corpus from NLTK using paras(). 

from nltk.corpus import brown

# Load Brown Corpus as paragraphs
brown_paragraphs = brown.paras()
```

```python
###2. Remove punctuation and stopwords. 

## Load the Brown Corpus as paragraphs
brown_paragraphs = brown.paras()

# Load the stop words for English
stop_words = set(stopwords.words('english'))

# Process paragraphs
processed_paragraphs = []

for paragraph in brown_paragraphs:
    # Flatten paragraph into a single text string
        text = ' '.join([' '.join(sentence) for sentence in paragraph]).lower()
    
# Remove punctuation
filtered_text = text.translate(str.maketrans('', '', string.punctuation))
    
# Tokenize the text
tokens = filtered_text.split()
    
# Remove stopwords
filtered_tokens = [word for word in tokens if word not in stop_words]
    
# Store the cleaned paragraph as a string
processed_paragraphs.append(' '.join(filtered_tokens))

# Print the first processed paragraph
print(processed_paragraphs[0])


```

```python
###3. Apply the lancaster stemmer. 

lancaster = nltk.LancasterStemmer()

lancaster_stems = [lancaster.stem(word) for word in filtered_tokens]

print(lancaster_stems)
```

```python
#4. Print to the screen the top 10 words in terms of TF. Show the TF values as well. 

tokens_stemmed = [nltk.LancasterStemmer().stem(token) for token in filtered_tokens]
tf = nltk.FreqDist(tokens_stemmed)
tf.most_common()[:20]
```

```python
#5. Print to the screen the top 10 words in terms of TF-IDF. Use the paragraphs as documents for calculating TF-IDF. Show the TF-IDF values as well.

# Initialize the TF-IDF dictionary
tf_idf = {}

# Parameters
doc_size = 1000
ntokens = len(filtered_tokens)
ndocs = math.ceil(ntokens / doc_size)

# Assuming `tf` contains the term frequencies for the words in the corpus
for token in tf:
    count = 0
    doc_idx = 0
    
    # Calculate how many documents contain the token
    while doc_idx < ndocs:
        if doc_idx == ndocs - 1:
            doc = filtered_tokens[doc_idx * doc_size:]
        else:
            doc = filtered_tokens[doc_idx * doc_size: (doc_idx + 1) * doc_size]
            
        # Check if the token is in the current document
        if token in doc:
            count += 1
            
        doc_idx += 1
    
    # Avoid division by zero: Check if count is zero
    if count > 0:
        tf_idf[token] = tf[token] * math.log(ndocs / count)
    else:
        # If the word doesn't appear in any document, we set its TF-IDF value to 0
        tf_idf[token] = 0

# Sort TF-IDF by value in descending order
sorted_tf_idf = sorted(tf_idf.items(), key=operator.itemgetter(1), reverse=True)

sorted_tf_idf[:20]
```

```python
#6. Use pos_tag() to tag each token. 

# use NLTK to tag after punctuation and stopword removal

tags = nltk.pos_tag(filtered_tokens)
tags[:20]
```

```python
#7. Print to the screen the 10 most common trigrams of word-tag pairs. Show their frequencies as well. Use nltk.trigrams().



# Load the tagged Brown corpus (word-tag pairs)
tagged_corpus = brown.tagged_sents()

# Generate trigrams of word-tag pairs using nltk.trigrams
trigrams = []

for sent in tagged_corpus:
    trigrams.extend(nltk.trigrams(sent))

# Count the frequency of each trigram
fdist = FreqDist(trigrams)

# Print the 10 most common trigrams with their frequencies
print("Top 10 Most Common Trigrams of Word-Tag Pairs:")
for trigram, freq in fdist.most_common(10):
    print(f"{trigram}: {freq}")
```
